```python
import pandas as pd
from sklearn import metrics

input_data = pd.DataFrame()
input_data['income']    = [60, 85.5, 64.8, 61.5, 87, 110.1, 108, 82.8, 69, 93, 51, 81, 75, 52.8, 64.8, 43.2, 84, 49.2, 59.4, 66, 47.4, 33, 51, 63]
input_data['plot_size'] = [18.4, 16.8, 21.6, 20.8, 23.6, 19.2, 17.6, 22.4, 20, 20.8, 22, 20, 19.6, 20.8, 17.2, 20.4, 17.6, 17.6, 16, 18.4, 16.4, 18.8, 14, 14.8 ]
input_data['owner']     = ['B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB', 'NB']
```

```python
input_data
```

```python
from sklearn import tree
ctree = tree.DecisionTreeClassifier() #instatiation
ctree
```

```python
x = input_data[['income', 'plot_size']]
y = input_data['owner']
x.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=1)
```

```python
ctree_model = ctree.fit(x_train, y_train)
```

```python
ctree_model
```

```python
y_train_pred = ctree_model.predict(x_train)
print(metrics.accuracy_score(y_train, y_train_pred))
```

```python
y_test_pred = ctree_model.predict(x_test)
print(metrics.accuracy_score(y_test, y_test_pred))
```

```python
comparision_df = x_test.copy()
comparision_df['actual'] = y_test
comparision_df['Test_predicted'] = y_test_pred
comparision_df
```

```python
ctree = tree.DecisionTreeClassifier(criterion='gini')
ctree = tree.DecisionTreeClassifier(max_depth=4) # depth = 1, 2, 3, 4 
ctree = tree.DecisionTreeClassifier(min_samples_split=3) # depth = 1, 2, 3, 4 

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4)

ctree_model = ctree.fit(x_train, y_train)

y_train_pred = ctree_model.predict(x_train)
print('train_accuracy: ', metrics.accuracy_score(y_train, y_train_pred))

y_test_pred = ctree_model.predict(x_test)
print('test_accuracy :', metrics.accuracy_score(y_test, y_test_pred))
```
